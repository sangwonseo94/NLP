```python
import os
import requests
from bs4 import BeautifulSoup

url = "https://search.naver.com/search.naver?&where=news&query=%EB%AC%B4%EC%97%AD%EC%A0%84%EC%9F%81&sm=tab_pge&sort=0&photo=0&field=0&reporter_article=&pd=0&ds=&de=&docid=&nso=so:r,p:all,a:all&mynews=0&cluster_rank=26&start=" 
news_dictionary = {}
url_list =[]

for k in range(1,110,10):
    url_list.append(url+str(k)+"&refresh_start=0")
    
for i in url_list:
    html = requests.get(i).text
    parser = BeautifulSoup(html,'lxml')
    plz_parsing = parser.select("dl > dt > a")
    title = [title.get('title') for title in plz_parsing]
    del title[0]
    url = [url.get('href') for url in plz_parsing]
    del url[0]
    for arr in range(0,len(url)):
        news_dictionary[title[arr]] = url[arr]
        
```
