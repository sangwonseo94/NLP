```python
import torch
lr= 0.01
X = torch.Tensor([[0,0],[0,1],[1,0],[1,1]])
Y = torch.Tensor([[0],[1],[1],[0]])

w1 = torch.Tensor(2,2)
w2 = torch.Tensor(2,1)
b1 = torch.Tensor(1)
b2 = torch.Tensor(1)

def sigmoid_(x):
    return 1.0 /(1.0 + torch.exp(-x))
def sigmoid_prime(x):
    return sigmoid_(x) * (1 -sigmoid_(x))

for step in range(10001):
    L1 = torch.add(torch.mm(X,w1),b1)
    act_1 = sigmoid_(L1)
    L2 = torch.add(torch.mm(act_1,w2),b2)
    y_prediction = sigmoid_(L2)
    
    cost = -torch.mean(Y * torch.log(y_prediction) + (1-Y)* torch.log(1-y_prediction))
    
    
    # backpropagation
    de_y_prediction = (y_prediction -Y) / (y_prediction * (1.0 - y_prediction) +1e-7)
    
    de_L2 = de_y_prediction * sigmoid_prime(L2)
    de_b2 = de_L2
    de_w2 = torch.mm(torch.transpose(act_1,0,1),de_b2)
    
    de_act1 = torch.mm(de_b2, torch.transpose(w2,0,1))
    de_L1 = de_act1* sigmoid_prime(L1)
    de_b1 = de_L1
    de_w1 = torch.mm(torch.transpose(X,0,1),de_b1)
    
    w1 = w1 - lr * de_w1
    b1 = b1 - lr * torch.mean(de_b1,0)
    w2 = w2 - lr * de_w2
    b2 = b2 - lr * torch.mean(de_b2,0)
    
    if step %100 == 0:
        print(step,cost.item())
```
