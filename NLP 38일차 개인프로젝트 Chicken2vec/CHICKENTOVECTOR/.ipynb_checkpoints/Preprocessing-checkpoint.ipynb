```python
import re
import pandas as pd
from konlpy.tag import Kkma
## 메뉴 뽑아오기 ##

stop_word_list = []
Chicken_menu_list= []
Chicken_review_list =[]
with open("stopwords.txt","r") as s:
    temp = s.read()
    stop_word_list = temp.split("\n")


## C2V에서 Name_list를 가져와서 작업 수행 ###

Chicken_dataframe= pd.read_csv("BHC-응암행복점.csv")


Chicken_menu = Chicken_dataframe['메뉴']

for Chicken_number in range (len(Chicken_menu)):
    
    if(str(Chicken_menu[Chicken_number]) == "nan"):
        Chicken_menu_list.append("")
    
    else:
        Chicken_name = Chicken_menu[Chicken_number].split("/")[0]
        Chicken_menu_list.append(Chicken_name)
```

```python
import re
p = re.compile(r"[^가-힣 ]+")
Chicken_review = Chicken_dataframe['리뷰']
Chicken_review_list = []
for review in Chicken_review:
    review= re.sub(p,"",review)
    review.strip()
    
    Chicken_review_list.append(review)
    
```

```python
result_DF = pd.DataFrame({"review":Chicken_review_list,"object":Chicken_menu_list})
for arr in range(len(result_DF)):
    if(result_DF["object"][arr] ==""):
        result_DF =result_DF.drop(arr)

result_DF
```

```python
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

### 모든 Name List를 Join 할 것 ### 우리동네에는 무슨치킨이 제일 많이 팔렸을까? ###

result_text = " ".join(result_DF['object'])
dak_mask = np.array(Image.open("dak.jpg"))

wordclo= WordCloud(background_color="white",mask=dak_mask,font_path="C:\Windows\Fonts\HMKMRHD.ttf",max_words=len(result_DF))
wordclo=wordclo.generate(result_text)

plt.figure(figsize=(15,15))
plt.imshow(wordclo, interpolation="bilinear")
plt.axis("off")
```

```python
from konlpy.tag import Okt  
okt = Okt()
sentence_list = []
### 전체 문장을 원형 추출을 한다 ###
for review_ in Chicken_review_list:
    Chicken_word= okt.pos(review_, stem=True, norm=True)
    temp = []
    for arr in Chicken_word:
        if(arr[0] not in stop_word_list):
            temp.append(arr[0])
    sentence_list.append(temp) 
sentence_list
len(sentence_list)
```

```python
## 원형 추출과 매핑되는 메뉴 Dataframe ###
Compare_Chicken_dataframe = pd.DataFrame({"review":sentence_list , "item": Chicken_menu_list})
Compare_Chicken_dataframe = Compare_Chicken_dataframe.dropna()
Compare_Chicken_dataframe
```

```python

## 전체 review의 list를 넣는다 ###

Total_review_list = Chicken_review_list

Chicken2Vec = []

for Review_one in Total_review_list:
    Chicken_word = okt.pos(Review_one, stem=True, norm=True)
    temp=[]
    for word in Chicken_word:
            if word[1] in ["Noun"]:
                temp.append((word[0])) 
            elif word[1] in ["Adjective"]:
                temp.append((word[0]))
    if temp: # 명사 형용사 원본 저장
          Chicken2Vec.append(temp) 
Chicken2Vec
```

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences=Chicken2Vec, size=100, window=5, min_count=5, workers=4, sg=0)
```

```python
keyword = input()
```

```python
search_base=model.wv.most_similar(keyword)
print(search_base)
```

```python
### a에서 명사 형용사 원본을 가져오는 부분 ###

search_base_list =[]
for search_base_item in search_base:
    search_base_list.append(search_base_item)
surgest_item = []
for compare_one in Compare_Chicken_dataframe["review"]:
    for search_base_item in search_base_list:
        if(search_base in compare_one):
            surgest_item.append(search_base_item)
            
set(surgest_item)
surgest_item
```
