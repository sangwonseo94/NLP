```
import pandas as pd
import requests
from bs4 import BeautifulSoup
# https://finance.naver.com/sise/sise_market_sum.nhn?&page=3
url = "https://finance.naver.com/sise/sise_market_sum.nhn?&page=1"


r = requests.get(url)
df = pd.read_html(r.text, header =0)
df[1].head(15).dropna(how='all')
```

```
url = 'https://finance.naver.com/sise/field_submit.nhn'
params = {
      'menu': 'market_sum',
  'returnUrl': 'http://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page=1',
  'fieldIds': [ # 27개 필드에서 선택(혹은 전체 사용)
    'quant', #거래량
    'open_val', #시가
    'high_val', #고가
    'low_val', #저가
    'buy_total', #매수총잔량
    'sell_total', #매도총잔량
    'amount', #거래대금(백만)
  ]

}


df_list = []
prev = ''
sosok = 0 
for page in range(1,50):
  params['returnUrl'] = 'http://finance.naver.com/sise/sise_market_sum.nhn?sosok=%d&page=%d' % (sosok, page)
  r = requests.get (url ,params = params)
  if prev== r.text : break
  prev= r.text
  df = pd.read_html(r.text)[1]
  df = df.dropna(how = "all")
  df_list.append(df)

result = pd.concat(df_list)
del result['N']
del result['토론실']
result.head(20)
```
