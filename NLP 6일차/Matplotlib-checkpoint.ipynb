```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline

x = np.arange(-2*np.pi,2*np.pi,0.1)
y1= np.cos(x)
y2= np.sin(x)
y3= np.tan(x)
y4= np.arctan(x)
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.ylim(-10,10)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```

```python
x = np.arange(-5,5)
y=  x**3
plt.plot(x,y)
[min_x,max_x] = plt.xlim()
print([min_x,max_x])

plt.grid()
plt.show()
```

```python
x = np.linspace(-4,4,200)
y1 = x**3
y2 = 10*x**2-2
y3 = 10*x**2
plt.plot(x,y1,'r--*',x,y2,'--*',x,y3,'*')
plt.ylim(-3,3 )
plt.xlim(-1,1)
plt.show()
```

```python
x = np.arange(0,5,1)
y1 = x
y2 = x +1
y3 = x +2
y4 = x +3
plt.plot(x,y1,'ro',x,y2,'g>',x,y3,'b<',x,y4,'cD')
plt.show()
```

```python
x = np.arange(-5,6,1)
y1 = x**2
y2 = x**2+1
y3 = x**2+2
y4 = x**2+3
plt.plot(x,y1,'-',x,y2,'--',x,y3,':',x,y4,'-.')

plt.xlim(-2,2)
plt.ylim(-1,12)
plt.legend(["data","data2","data3","data4"],loc="best")
plt.show()
```

```python
%matplotlib inline
x = np.arange(-4.5,5,0.5)
y1 = 2*x**3
y2 = 3*x**3
plt.plot(x,y1,'r--',x,y2,"b--")
plt.xlabel("x축")
plt.ylabel("y축")
plt.title("그래프")
plt.text(2,200,"이것은 그래프")
plt.legend(["y=2x^3","3x^3"],loc='best')
plt.show()
```
