```python
from selenium import webdriver
import time
from os import listdir
import pandas as pd
def click(path):
    element = webdriver.find_element_by_xpath(path)
    webdriver.execute_script("arguments[0].click();", element)
def Crawling_Target():
    Total_name_for_Crawling= []
    Folder_Path = 'C:\레시피크롤링\분할'
    Folder_list = [arr for arr in listdir(Folder_Path)]
    for One_Folder in Folder_list:
        File_Path = Folder_Path + "\\" + One_Folder
        File_list = [arr for arr in listdir(File_Path)]
        for arr in File_list:
            df = pd.read_excel(File_Path+'\\'+arr)
            name = list(df.iloc[:,0])
            for unique_name in name:
                unique_name  = unique_name.replace("("," ")
                unique_name = unique_name.replace(")","")
                unique_name = unique_name.replace("&"," ")
                Total_name_for_Crawling.append(unique_name)
    return list(set(Total_name_for_Crawling))
```

```python
from selenium import webdriver
webdriver = webdriver.Chrome("C:\레시피크롤링\chromedriver")
webdriver.get('http://www.10000recipe.com/')
```

```python
Food ={}
Basic =1
bool_next = 0
booling_next = False
for index in range(0,len(CT)):
    
    if(bool_next ==1):
        index -=1
        Basic +=1
    if(bool_next >=2):
        Basic +=1
    webdriver.find_element_by_xpath('//*[@id="srhRecipeText"]').send_keys(CT[index])
    ## key 부분에 메뉴가들어간다.
    try:
        webdriver.find_element_by_xpath('//*[@id="frmTopRecipeSearch"]/div/span/button/span').click()
        click('//*[@id="contents_area_full"]/div/div[3]/div[2]/div['+str(Basic)+']/a/span/img')
        step = []
    except:
        step = []
        Food[CT[index]] = step
        없는메뉴.append(CT[index])
        click("/html/body/div[1]/div[1]/div/h1/a/img")
        continue
    for i in range(1,100):
        try:
            temp = []
            temp.append(i)
            temp.append(webdriver.find_element_by_xpath('//*[@id=\"stepdescr' + str(i) +'\"]').text)
            try:
                temp.append(webdriver.find_element_by_xpath('//*[@id=\"stepimg' + str(i)+'\"]/img').get_attribute("src"))
            except:
                if(bool_next >=1):
                    bool_next +=1
                    booling_next = True
                    break
                bool_next = 1
                booling_next =True
                break
            step.append(temp)
        except:
            booling_next=False
            break
    if(booling_next):
        click("/html/body/div[1]/div[1]/div/h1/a/img")
        continue
    booling_next = False
 
    
    ingredients = webdriver.find_element_by_xpath('//*[@id="divConfirmedMaterialArea"]').text
    ingredients = ingredients.replace("\n"," ")
    if ( len(ingredients) == 0):
        print(webdriver.find_element_by_xpath('//*[@id="contents_area"]/div[2]/h3').text + "재료가없습니다")
    
    name = webdriver.find_element_by_xpath('//*[@id="contents_area"]/div[2]/h3').text
    Food[CT[index]] = [ [ingredients , name]  , step]
    bool_next =0
    Basic = 1
    click("/html/body/div[1]/div[1]/div/h1/a/img")
```
