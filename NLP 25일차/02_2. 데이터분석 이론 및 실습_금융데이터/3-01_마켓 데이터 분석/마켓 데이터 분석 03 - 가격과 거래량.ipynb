# 2016년 특정 종목 데이터 

```python
start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2016, 12, 31)

# 셀트리온(068270)
df = fdr.DataReader("068270", start, end)
df.head()
```

# 거래량 Volume

1. 일반적으로 거래량은 주가변화에 선행
1. 일반적으로 거래량이 늘면 주가는 상승하고 거래량이 줄면 주가는 하락
1. 거래량에 급격한 변화가 있으면 주가가 급등하거나 급락

```python
df['Volume'].plot(color='g')
```

# 거래가 없는 ROW 삭제 

```python
df = df[df.Volume > 0]
df.head()
```

# 가격

```python
df['Close'].plot(figsize=(16, 4), title='005930 Price 2016')
```

# 거래량

```python
df['Volume'].plot(figsize=(16, 4), style='g', title='005930 Volume 2016')
```

## 거래량과 주가비교

```python
df2 =  df[['Close', 'Volume']]
df2.head()
```

```python
df2.plot()
```

## 두 번째 Y축 Secondary Y

```python
df2['Close'].plot(figsize=(16, 4), style='b')
df2['Volume'].plot(figsize=(16, 4), style='g', secondary_y=True)
```

```python
df2['2016-07':]['Close'].plot(style='b')
df2['2016-07':]['Volume'].plot(style='g', secondary_y=True)
```

# 정규화 (normalization)
데이터의 변화를 상대적으로 비교

```python
df2.mean()
```

```python
df_norm = (df2 - df2.mean()) / (df2.max() - df2.min())
df_norm.head()
```

```python
df_norm.plot()
```
