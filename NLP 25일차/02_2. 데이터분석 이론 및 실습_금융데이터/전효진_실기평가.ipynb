과거 2014-01-01 부터 2018-12-31 까지 애플(AAPL)종목의 가격 데이터와
S&P500 지수 데이터를 가져와 하나의 차트를 그려보고, 수익률을 비교

```python
%matplotlib inline
import pandas as pd
import FinanceDataReader as fdr
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (14,8)
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["axes.grid"] = True
plt.rcParams['axes.axisbelow'] = True
```

```python
sp500 = fdr.StockListing('S&P500')
aapl = fdr.DataReader('AAPL', '2014-01-01', '2018-12-31', country='US')
us500 = fdr.DataReader('US500', '2014-01-01', '2018-12-31')

df_plot = pd.DataFrame()

df_plot['S&P500'] = us500['Close']
df_plot['AAPL'] = aapl['Close']

df_plot = df_plot / df_plot.iloc[0] - 1.0
df_plot.plot()
```
