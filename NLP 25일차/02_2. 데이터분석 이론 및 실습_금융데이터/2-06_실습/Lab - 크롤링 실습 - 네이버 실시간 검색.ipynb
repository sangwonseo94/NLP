# 네이버 실시간 검색 순위

```python
import json, requests
from pandas.io.json import json_normalize

r = requests.get('http://rank.search.naver.com/rank.js')
df = json_normalize(json.loads(r.text), ['data', 'data'])
df.head()
```

# 엑셀로 저장


```python
df.to_excel('real_search.xlsx')
```
