# 네이버 파이낸스 - 시가총액 팩터 데이터

```python
url = 'https://finance.naver.com/sise/field_submit.nhn' 
  
params = {
  'menu': 'market_sum',
  'returnUrl': 'http://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page=1',
  'fieldIds': ['quant', 'market_sum', 'operating_profit', 'sales'],
}

r = requests.get(url, params=params)
dfs = pd.read_html(r.text, header=0)
df = dfs[1]
df.head(15)
```

# 페이지 네비게이션
매 페이지를 DataFrame으로 얻어, DataFrame들을 병합(concat)

```python
url = 'https://finance.naver.com/sise/field_submit.nhn'
params = {
  'menu': 'market_sum',
  'returnUrl': 'http://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page=1',
  'fieldIds': [ # 27개 필드에서 선택(혹은 전체 사용)
    'quant', #거래량
    'open_val', #시가
    'high_val', #고가
    'low_val', #저가
    'buy_total', #매수총잔량
    'sell_total', #매도총잔량
    'amount', #거래대금(백만)
  ]
}

# df 리스트
df_list = []
prev_resp = ''
sosok = 0 # 0=코스피, 1=코스닥

for page in range(1, 50): # 최대 49페이지를 넘지 않을 것으로 가정
    params['returnUrl'] = 'http://finance.naver.com/sise/sise_market_sum.nhn?sosok=%d&page=%d' % (sosok, page),
    r = requests.get(url, params=params)
    if prev_resp == r.text: # 바로 이전 페이지와 내용이 같으면 중단
      break
    prev_resp = r.text
    df = pd.read_html(r.text)[1]
    df = df.dropna(how='all')
    df_list.append(df) # df_list에 페이지별 df추가

# df_list의 모든 df를 연결
result = pd.concat(df_list)
```

```python
result = result.reset_index(drop=True)
del result['토론실']
del result['N']

result.head(20)
```

## 함수로 정리하기
코스피, 코스닥 포함 전체(모든 페이지) 취합

```python
import requests
import pandas as pd
import urllib
import io

def naver_marcap_factors():
  url = 'https://finance.naver.com/sise/field_submit.nhn'
  params = {
    'menu': 'market_sum',
    'returnUrl': 'http://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page=1',
    'fieldIds': [ # 27개 필드에서 선택(혹은 전체 사용)
      'quant', #거래량
      'open_val', #시가
      'high_val', #고가
      'low_val', #저가
      'ask_buy', #매수호가
      'ask_sell', #매도호가
      'buy_total', #매수총잔량
      'sell_total', #매도총잔량
      'amount', #거래대금(백만)
      'prev_quant', #전일거래량
      'frgn_rate', #외국인비율
      'listed_stock_cnt', #상장주식수(천주)
      'market_sum', #시가총액(억)
      'property_total', #자산총계(억)
      'debt_total', #부채총계(억)
      'sales', #매출액(억)
      'sales_increasing_rate', #매출액증가율
      'operating_profit', #영업이익(억)
      'operating_profit_increasing_rate', #영업이익증가율
      'net_income', #당기순이익(억)
      'eps', #주당순이익(원)
      'dividend', #보통주배당금(원)
      'per', #PER(배)
      'roe', #ROE(%)
      'roa', #ROA(%)
      'pbr', #PBR(배)
      'reserve_ratio', #유보율(%)
    ]
  }

  # df 리스트
  df_list = []
  prev_resp = ''
  for sosok in [0, 1]: # 0=코스피, 1=코스닥
    for page in range(1, 50): # 최대 49페이지를 넘지 않을 것으로 가정
        params['returnUrl'] = 'http://finance.naver.com/sise/sise_market_sum.nhn?sosok=%d&page=%d' % (sosok, page),
        r = requests.get(url, params=params)
        if prev_resp == r.text: # 바로 이전 페이지와 내용이 같으면 중단
          break
        prev_resp = r.text
        df = pd.read_html(r.text)[1]
        df = df.dropna(how='all')
        df['시장'] = '코스피' if sosok==0 else '코스닥'
        df_list.append(df) # df_list에 페이지별 df추가

  # df_list의 모든 df를 연결
  result = pd.concat(df_list)
  result = result.reset_index(drop=True)
  del result['토론실']
  del result['N']
  
  # 등락률 타입 변환
  result['등락률'] = result['등락률'].str.replace('%', '').astype(float) 
  return result
```

```python
df_rank = naver_marcap_factors()
df_rank.head()
```

## 종목 발굴

1. 시가총액 상위 500개 종목
1. 영업이익증가율 상위 100개 종목 
1. 시가총액 영업이익 비율 (코스피 20% 이상, 코스닥 10% 이상)

참고:
* 성장성 중요도: 매출액 증가 < 당기순이익 증가 < 영업이익 증가
* 시가총액 영업이익 비율 = 시가총액 / 영업이익

```python
df_rank['시가총액_영업이익_비율'] = df_rank['영업이익'] / df_rank['시가총액']
```

```python
df_rank.nlargest(500, '시가총액').nlargest(100, '영업이익증가율').query('시가총액_영업이익_비율 > 0.2 and 시장=="코스피"').sort_values('시가총액', ascending=False)
```

```python
df_rank.nlargest(500, '시가총액').nlargest(100, '영업이익증가율').query('시가총액_영업이익_비율 > 0.1 and 시장=="코스닥"').sort_values('시가총액', ascending=False)
```
